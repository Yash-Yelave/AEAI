```python
# Install:
!pip install langchain langchain-community huggingface_hub

from langchain_community.llms import HuggingFaceHub
from langchain_core.prompts import PromptTemplate
from langchain.chains import LLMChain

# 🔑 Add your FREE HuggingFace API key
import os
os.environ["HUGGINGFACEHUB_API_TOKEN"] = "your_token_here"

# Load free model
llm = HuggingFaceHub(
    repo_id="google/flan-t5-large",   # free & good for reasoning
    model_kwargs={"temperature": 0.5, "max_length": 256}
)

# Prompt with explainability
prompt = PromptTemplate(
    input_variables=["question"],
    template="""
You are an explainable AI assistant.

Answer the question AND explain reasoning step-by-step.

Format:
Answer: <final answer>
Reasoning:
1. Step 1
2. Step 2
3. Step 3

Question: {question}
"""
)

# Chain
chain = LLMChain(llm=llm, prompt=prompt)

# Chat loop
while True:
    q = input("You: ")
    if q.lower() == "exit":
        break

    response = chain.run(q)
    print("\nBot:\n", response)
```
