```python
import queue
import time

class Message:
  def __init__(self, sender, receiver, content):
    self.sender = sender
    self.receiver = receiver
    self.content = content

class WorkerAgent:
  def __init__(self, name):
    self.name = name

  def receive_task(self, message):
    task = message.content
    print(f"[{self.name}] Receved Task from {message.sender}: {task}")

    time.sleep(1)

    result = f"{self.name} Completed: {task}"

    return Message(self.name, message.sender, result)

class ManagerAgent:
  def __init__(self):
    self.name = 'Manager'
    self.workers = [
        WorkerAgent('Agent_A'),
      WorkerAgent('Agent_B'),
      WorkerAgent('Agent_C')
    ]

    self.task_queue = queue.Queue()

  def decompose_task(self, task):
    return task.split(" and ")

  def delegate_task(self, task):
    print(f"[{self.name}] Received Task:{task}")

    subtasks = self.decompose_task(task)

    for sub in subtasks:
      self.task_queue.put(sub)

    results = []
    workers_index = 0

    while not self.task_queue.empty():
      subtask = self.task_queue.get()

      worker = self.workers[workers_index % len(self.workers)]
      workers_index += 1

      print(f"[{self.name}] Delegating '{subtask}' to {worker.name}")

      message = Message(self.name, worker.name, subtask)

      response = worker.receive_task(message)

      print(f"[{self.name}] Received result: {response.content}")
      results.append(response.content)

    return results


if __name__ == "__main__":
  manager = ManagerAgent()

  task = "Collect Data and Process data and train model and evaluate model"

  final_results = manager.delegate_task(task)

  print("Final Combintion Results:")
  for res in final_results:
    print(res)
```
